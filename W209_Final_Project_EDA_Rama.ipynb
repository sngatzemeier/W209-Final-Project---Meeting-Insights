## 1. Introduction
Collaboration has a significant value by which it enables organizations to achieve their goals. In any organization, all jobs are dependent on cross-functional teams of individuals coming together to address the challenge they face collectively. Besides, collaboration spurs innovation. Such collaboration happens in various settings (examples include one-to-one discussion, brainstorming sessions, team discussions, ad-hoc meetings, structured updates) pursued by different organizations and individuals. All these forums are focused on bringing people together to share their ideas and solve problems. In the post-COVID-19 pandemic effect, even an informal talk is a zoom meeting. There is an increasing number of meetings that occupy our calendars. While these meetings are essential, an overwhelming number of them can be energy-draining and prevent individuals from analyzing time. Again, not all meetings have to be efficient and effective. Conferences are a means to an end, helping individuals achieve assigned goals. One has to wonder if the sessions are efficient and enable us to achieve our goals. Besides, managers and individuals prioritize how they spend their time on topics to get their valuable time slots. This project aims to provide visualizations that help individuals and managers achieve the goal of understanding how their time is utilized during a workday and how it can be streamlined to better their chances of success

## Hypothesis 1: 
## More than 50% of the meetings are not effective

**Assumption made:** 
A meeting is productive when most of the invitees attend the meeting, they are well prepared for the topic and provide direction or contribute to solutions

**Observation:**
*Section 9* has the data analysis performed to assess attendees' productivity 

The chart in [Tableau Report 5](https://public.tableau.com/profile/ramakrishna.rangavajjula#!/vizhome/AttendeeProductivity/AttendeeProductivity?publish=yes) clearly shows:

i) Over 90% of people attended the meetings. Therefore, most invitees attended

ii) Between 30-40% of people had meeting conflicts. Therefore, they didn't manage their time effectively and were probably not well prepared for the meeting

iii) Between 50-60% of people were multitasking during meetings, therefore, were not entirely contributing/giving direction in the meeting

iv) Between 50-60% of people were drafting emails during meetings, therefore, were not entirely /giving direction in the meeting

## Hypothesis 1 Conclusion

Meeting attendance from the invitees was very high however a large percentage didn't prepare for the meetings, and more than 50% of them were distracted with others' tasks and drafting emails. Therefore, the hypothesis is true



## Hypothesis 2:
## Early morning and late night meetings are impacting work/life balance

**Assumption made:** People working from 9:00 AM to 5:00 PM local standard time and have no work over weekends have a right work/life balance

**Observation 1:** *Section 7* analyzed the data and found that individual Friday meetings extend into Saturdays. The chart in [Tableau Report 2](https://public.tableau.com/profile/ramakrishna.rangavajjula#!/vizhome/MeetingDurationHoursbyDay/MeetingDurationbyDay?publish=yes)
 clearly shows:

During  Feb' 20, Mar' 20, Nov' 20, Dec' 20, Jan' 21, and Feb' 21, meetings were either (a) organized over weekends or (b) late Friday night meetings extended into early morning Saturdays

**Observation 2:** *Section 10* analyzed the data on when meetings were being organized. The chart in [Tableau Report 6](https://public.tableau.com/profile/ramakrishna.rangavajjula#!/vizhome/MBTOD/MeetingsbyTimeofDay?publish=yes) clearly shows:

i) Meetings are not scheduled within the regular working hours of 9:00 AM PST to 1:30 PM CST

ii) Meetings are organized late during the regular working hours and beyond working hours until 1:00 AM 

## Hypothesis 2 Conclusion
Meetings are organized from about 1:30 PM local standard time to 1:00 AM. Additionally, individual Friday meetings weeks also extend into Saturday. Therefore, the ** Assumption made** is broken, and the Hypothesis is true
